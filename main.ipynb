<h2 style="color: #c333aa; text-decoration: underline">I. Introduction</h2>
<p>
    The aim is to <strong style="color: #c333aa">search</strong> for a source of news titles given a list of queries. The programme is simply built using <strong style="color: #c333aa">indexing</strong> on the source for a speedy performance. The source file is firstly indexed, followed by <strong style="color: #c333aa">row by row</strong> queries reading from the query file. The results are finally listed in order according to each row of query, each with ascending order of Title ID.
</p><p>
    The source file is <code>main.py</code> with the required CLI arguments <code>query</code>, <code>source</code> & <code>output</code> supported. A <code>TextSearch</code> object is firstly created with the given source file name as the parameter. During the read-in of source file, the data is being <strong style="color: #c333aa">indexed</strong>. The, the query file is read line by line, and each row of query is passed to <code>TextSearch.query()</code> for <strong style="color: #c333aa">query matching</strong>. This returns a <code>list</code> of matched title IDs, and stored to another result <code>list</code>, which finally gathers all rows of query results and output the results as a file.
</p>
<p>Noted that <strong style="color: #c333aa">no third party</strong> libraries is used.</p>

<h2 style="color: #c333aa; text-decoration: underline">III. Query Matching</h2>
<p>
    Each row of queries would need to be <strong style="color: #c333aa;">matched</strong> with the source titles for our targeted results. Firstly, the query row is <strong style="color: #c333aa;">splitted</strong> according to the defined query formats: <strong>" or "<strong>, </strong>" and "</strong>, or <strong>" not "</strong>. </p>
<p>
    For each of the word splitted, it will <strong style="color: #c333aa;">search</strong> the <strong>speed-related</strong> <code>dict</code> only using the <strong style="color: #c333aa;">first character</strong>. This returns a <code>set</code> of Title IDs which contains the first character of the query word.<br />
    For each IDs returned, it will then loop through the <strong>data-related</strong> <code>dict</code> using these IDs as the key. For each titles, it will then match with the <strong style="color: #c333aa;">full query word</strong> using native string matching <code>in</code> operator. As a results, for each query word, there will be a so-called <strong style="color: #c333aa;">filtered</strong> <code>set</code> of matched Title IDs.
</p>
<p>
    To come with the query operator <strong>" or "</strong>, <strong>" and "</strong>, or <strong>" not "</strong>, this is just like a <strong style="color: #c333aa;">set operation</strong>. From results above, we have several <code>set</code>s of matched titles. If the query operator is <strong>" or "</strong>, the <code>set</code>s are then taken under native <code>set</code> union operation (<code>|</code>). This is similar for <strong>" and "</strong> using <code>&</code>.<br />As for <strong>" not "</strong>, this is actually a <strong style="color: #c333aa;">complement</strong> operation (<code>-</code>), using the <code>set</code> of the result from the first query word, removing the union set of the <code>set</code>s of later query words.
</p>
<p>
    Using the simplified data in Section II, suppose we want to query "很帥 and 你很" (Let's just consider them as a valid vocab). Firstly, "很" in "很帥" will return a matched <code>set</code> of <code>{2,3,4}</code>, while "你" in "你很" will return a matched <code>set</code> of <code>{1,3}</code>. The <strong>" and "</strong> query operator infers a <strong style="color: #c333aa;">set intersection operation</strong>, so <code>{2,3,4}&{1,3}</code> computes the <code>{3}</code> as the contextual result of this row of query.
</p>,
